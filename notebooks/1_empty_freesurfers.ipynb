# Manage empty freesurfers

After the data download step (either by using the scripts provided or the first notebook), some patients may have not passed OASIS quality test. Those patients are recognizable by empty folders, that we want to separate from the full ones.

- Import necessary libraries

```python
import os
import shutil
import sys

# Append the path to access custom modules
if '../' not in sys.path:
    sys.path.append('../')
    
import src.utils.folder_management as fm
```

- Set the various path needed by the clean method

```python
freesurfers = os.path.join('../', 'data', 'freesurfers')
empty = os.path.join('../', 'data', 'empty')
mri = 'mri'
```

- Get the DataManager

```python
manager = fm.DataManagerMRI(freesurfer_output_dir=freesurfers)
```

- Move empty freesurfer to an apposite folder

```python
manager.clean_freesurfers(
    base_empty_dir=empty,
    mri_dir=mri
)
```

```python
print(f"Number of empty folders: {len(os.listdir(empty))}")
```

```python
print(f"Number of full folders: {len(os.listdir(freesurfers))}")
```
