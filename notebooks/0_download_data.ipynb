# Download data with DataManager

*Note: this notebook should be run ONLY if you decide to not use the scripts to download data* 

- Importing necessary library

```python
import sys
import os

# Append the path to access custom modules
if '../' not in sys.path:
    sys.path.append('../')
```

- Import the module that contain the DataManager

```python
import src.utils.data_manager_fs as fm
```

- Instantiate the manager object

```python
manager = fm.FreesurfersManagerMRI(freesurfer_output_dir=os.path.join('../', 'data', 'try'))
```

- Start the download (it may take a very very long time)

```python
manager.download(
    subjects_file_path=os.path.join('..', 'data', 'csv', 'fsublist.csv'),
    username=os.environ['nitrc_username'],
    password=os.environ['nitrc_password'],
    files_to_keep=['posterior_Right-Hippocampus.mgz', 'posterior_Left-Hippocampus.mgz']
)
```

- Once the download is completed, some empty folder may be present
- Sometimes it looks like the server does not recognize your NITRC credentials. I personally solved this problem multiple times by chaging NITRC account password.
